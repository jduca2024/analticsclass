# Importing Data and Pandas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv',index_col=0)
```

# Viewing, Cleaning Data, Updating Tables by changing data to numeric

```python
df.head(75)
```

```python
df.columns
```

```python
df.replace({'s': np.nan, 'na': np.nan}, inplace=True)
df['Mean Score'] = pd.to_numeric(df['Mean Score'], errors='coerce')
feature1 = 'Total Tested'
feature2 = 'Mean Score'
```

```python
df_cleaned = df.dropna(subset=[feature1])
df_cleaned2 = df_cleaned.dropna(subset=[feature2])

df_cleaned2.head()
```

```python
All_High_schools = df_cleaned2[df_cleaned2['School Level'] == 'High school']

abraham_lincoln_df = All_High_schools[All_High_schools['School Name'] == 'Abraham Lincoln High School']
```

# Get averages of All highschools, and singling out Abraham Lincoln High School

```python
All_High_schools = df_cleaned2[df_cleaned2['School Level'] == 'High school']

abraham_lincoln_df = All_High_schools[All_High_schools['School Name'] == 'Abraham Lincoln High School']

abraham_lincoln_score_mean = abraham_lincoln_df['Mean Score'].mean()

abraham_lincoln_total_tested = abraham_lincoln_df['Total Tested'].sum()

abraham_lincoln_score_mean = round(abraham_lincoln_score_mean,2)

All_High_schools_score_mean = All_High_schools['Mean Score'].mean()

All_High_schools_total_tested = All_High_schools['Total Tested'].sum()

All_High_schools_score_mean = round(All_High_schools_score_mean,2)

print("Average Mean Score for Abraham Lincoln High School:",abraham_lincoln_score_mean)
print("Total Tested for Abraham Lincoln High School:",abraham_lincoln_total_tested)
print("Average Mean Score for All High Schools:",All_High_schools_score_mean)
print("Total Tested for All High Schools:",All_High_schools_total_tested)
```

We then wanted to show the breakdown by average scores of tests for abraham lincoln and then compare to the average of tests by other schools.

```python
#group by for Regents and average score and Abraham Lincoln Average Score to compare

abraham_lincoln_exam_mean = abraham_lincoln_df.groupby('Regents Exam')['Mean Score'].mean().reset_index()
abraham_lincoln_exam_mean = round(abraham_lincoln_exam_mean,2)
print("Average Mean Score per Exam for Abraham Lincoln High School:",abraham_lincoln_exam_mean)
All_high_school_exam_mean = All_High_schools.groupby('Regents Exam')['Mean Score'].mean().reset_index()
All_high_school_exam_mean = round(All_high_school_exam_mean,2)
print("Average Mean Score per Exam for All Schools:",All_high_school_exam_mean)

```

# The normal scatter plot wasn't the best form of graph to use I used a variation of a bar chart.

```python
bar_width1 = 0.35
bar_width2 = 0.70

plt.figure(figsize=(12, 6))

bars1 = plt.bar(All_high_school_exam_mean['Regents Exam'], All_high_school_exam_mean['Mean Score'], bar_width2,color='orange', label='All High Schools')

bars2 = plt.bar(abraham_lincoln_exam_mean ['Regents Exam'], abraham_lincoln_exam_mean['Mean Score'], bar_width1,color='blue', label='Abraham Lincoln High School')
plt.legend()

for bar in bars1:
   yval = bar.get_height()
   plt.text(bar.get_x() + bar.get_width() / 2, yval, round(yval, 2), va='bottom', ha='center', color='black')
for bar in bars2:
   yval = bar.get_height()
   plt.text(bar.get_x() + bar.get_width() / 2, yval, round(yval, 2), va='bottom', ha='center', color='black')

plt.ylim(50, None)
plt.title('Comparison of Mean Score per Regents Exam: Abraham Lincoln High School vs All High Schools')
plt.xlabel('Regents Exam')
plt.ylabel('Average Mean Score')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# Summary: In summation, Abraham Lincoln high school scores very well in Algebra 2 and Trig but very very poorly in English and also in Physical Settings/Physics
